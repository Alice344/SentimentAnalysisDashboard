```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

data = pd.read_csv('file.csv')

# Create word clouds for positive, neutral, and negative sentiment
positive_text = " ".join(data[data['labels'] == 'good']['tweets'].astype(str).tolist())
negative_text = " ".join(data[data['labels'] == 'bad']['tweets'].astype(str).tolist())
neutral_text = " ".join(data[data['labels'] == 'neutral']['tweets'].astype(str).tolist())

positive_wordcloud = WordCloud(
    width=800,
    height=800,
    background_color='white',
    colormap='Greens',
    collocations=False
).generate(positive_text)

negative_wordcloud = WordCloud(
    width=800,
    height=800,
    background_color='white',
    colormap='Reds',
    collocations=False
).generate(negative_text)

neutral_wordcloud = WordCloud(
    width=800,
    height=800,
    background_color='white',
    colormap='Blues',
    collocations=False
).generate(neutral_text)

fig, ax = plt.subplots(1, 3, figsize=(18, 6))  

ax[0].imshow(positive_wordcloud, interpolation='bilinear')
ax[0].axis("off")
ax[0].set_title("Positive Sentiment")

ax[1].imshow(neutral_wordcloud, interpolation='bilinear')  
ax[1].axis("off")
ax[1].set_title("Neutral Sentiment")

ax[2].imshow(negative_wordcloud, interpolation='bilinear')  
ax[2].axis("off")
ax[2].set_title("Negative Sentiment")

plt.tight_layout() 
plt.show()
```
